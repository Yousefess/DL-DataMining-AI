```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split

import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import plot_model

import warnings
warnings.simplefilter('ignore')
```

```python
m = 200
X, y = datasets.make_classification(n_samples=m,
n_features=2,
n_informative=2,
n_redundant=0,
n_repeated=0,
n_classes=2,
n_clusters_per_class=2)
plt.scatter(X[:,0],X[:,1],c=y,marker='o');

y = y.reshape(-1,1)

n_inputs = X.shape[1]
n_outputs = 1
learning_rate = 0.01
n_epochs = 500
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
model = Sequential()
output_layer = Dense(units=1, input_dim=2, activation='sigmoid', name="output_layer")
model.add(output_layer)
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
model_history = model.fit(x_train, y_train, epochs=n_epochs)
```

```python
loss_values = model_history.history['loss']
accuracy_values = model_history.history['accuracy']
```

```python
plt.figure()
plt.plot(loss_values);
```

```python
plt.figure()
plt.plot(accuracy_values);
```

```python
model.evaluate(x_test, y_test)
```
